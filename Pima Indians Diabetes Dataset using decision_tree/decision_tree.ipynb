```python
#import Library
import pandas as pd
from sklearn import tree
```

```python
df = pd.read_csv('pima-indians-diabetes.csv')
```

```python
df
```

```python
# here I separated the label column from this data_set.
x = df.iloc[:,:-1]
```

```python
x
```

```python
# here I separated the features column from this data_set.
y  = df.iloc[:,8]
```

```python
y
```

# Apply Decision Tree Classifier

```python
classify_ = tree.DecisionTreeClassifier()
```

# Traning Data

```python
classify_ = classify_.fit(x,y)
```

# Predict or testing a value

```python
predict_ = classify_.predict([[9,170,74,31,0,44,0.403,43]])
```
